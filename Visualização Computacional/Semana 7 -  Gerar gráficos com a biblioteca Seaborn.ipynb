```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
url = 'https://github.com/higoramario/univesp-com400-visualizacao-computacional/raw/main/com400-semana-7-disciplinas-notas.csv'
notas = pd.read_csv(url)
```

```python
notas.head()
```

```python
len(notas)
```

```python
sb.histplot(data=notas, x='idade', bins=47)
plt.show()
```

```python
sb.countplot(x='sexo', data=notas, palette='colorblind')
plt.show()
```

```python
sb.violinplot(data=[notas['com310-nota'],notas['com400-nota']]).set_xticklabels(['COM310','COM400'])
plt.show()
```

```python
sb.jointplot(data=notas, y='idade',x='com400-nota', kind='kde')
plt.show() 
```
