```python
import pandas as pd
import plotly.express as px
```

Pegando os dados da OD17 para o exercício.

```python
url = 'https://github.com/higoramario/univesp-com400-visualizacao-computacional/raw/main/com400-semana-6-od17.csv'
od17 = pd.read_csv(url)
od17.head()
```

Gerando o gráfico de sunburst

```python
sunburst = px.sunburst(od17,path=['MODOPRIN','CRITERIOBR'], values='FE_VIA')
sunburst.show()
```

Gerando o mapa de árvore

```python
arvore = px.treemap(od17,path=['MODOPRIN','CRITERIOBR'], values='FE_VIA')
arvore.show()
```
