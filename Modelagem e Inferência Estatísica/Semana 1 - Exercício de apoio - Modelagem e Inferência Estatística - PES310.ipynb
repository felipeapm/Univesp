Neste caso a variável Motor apenas é usada para identificar a amostra, portanto não é relevante neste momento (iria ser relevante se precisarmos por exemplo evitar usar esse tipo de motor por causa de algum resultado obtido na análise).

Com esses dados, a relação entre a idade do motor do cortador de grama e suas emissões de NOx, tanto com gasolina comum quanto com gasolina reformulada, parece um tanto duvidosa. Essa observação se confirma mediante os valores de correlação calculados, os quais apresentam uma correlação moderada.
Seria de se esperar que quanto maior a idade do cortador de grama as emissões também aumentariam. Nos diagramas não é possível observar esse padrão. A princípio pode-se inferir que a idade do motor não parece ser um preditor útil da emissão de NOx.

# Exercício 3
Os dados a seguir sobre y = concentração de amônio (mg/L) e x = transpiração (ml/h) foram lidos a partir de um gráfico no artigo: Resposta da remoção de amônio ao crescimento e transpiração de juncus efusus durante o tratamento de esgoto artificial em áreas úmidas em escala de laboratório (*Response of ammonium removal to growth and transpiration of juncus effusus during the treatment of artificial sewage in laboratory-scale wetlands*, Water Research, 2013: 4265-4273). 
O resumo do artigo declarou que “uma correlação linear entre a concentração de amônio dentro da rizosfera e a transpiração das plantas implica que existe uma influência da atividade fisiológica da planta na eficiência da remoção de N”. (A rizosfera é a região estreita do solo na interface raiz-solo da planta e a transpiração é o processo de movimento da água através de uma planta e sua evaporação.) 
O artigo relatou quantidades resumidas de uma análise de regressão linear simples. Com base em um diagrama de dispersão, como você descreveria a relação entre as variáveis e a regressão linear simples parece ser uma estratégia de modelagem adequada? 
Para responder observe os dados obtidos do artigo.

```python
#@title Tabela de Dados
from tabulate import tabulate
table = [["x", 5.8, 8.8, 11.0, 13.6, 18.5, 21.0, 23.7,
          26.0, 28.3, 31.9, 36.5, 38.2, 40.4],
         ["y", 7.8, 8.2, 6.9, 5.3, 4.7, 4.9, 4.3,
          2.7, 2.8, 1.8, 1.9, 1.1, 0.4]]
print(tabulate(table))
```

## Resposta 
Inicialmente é necessário realizar um gráfico de dispersão dos dados

```python
# definir dados
x = [5.8, 8.8, 11.0, 13.6, 18.5, 21.0, 23.7,
          26.0, 28.3, 31.9, 36.5, 38.2, 40.4] # transpiração (ml/h)
y = [ 7.8, 8.2, 6.9, 5.3, 4.7, 4.9, 4.3,
          2.7, 2.8, 1.8, 1.9, 1.1, 0.4] #  concentração de amônio (mg/L) 
```

```python
#@title Diagrama de dispersão: Remoção de N
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#plotar e calcular a correlação
plt.scatter(x, y, label=f'Correlação = {np.round(np.corrcoef(x,y)[0,1], 2)}')
plt.xlabel('Transpiração [ml/h] ')
plt.ylabel('Concentração de NH₄⁺ [mg/L] ') 
plt.title('Remoção de N')
plt.legend()
plt.show()
```

Pode-se observar que no diagrama de dispersão que os dados tendem a ficar próximos de uma reta com tendência decrecente, portanto é possível afirmar que a correlação é negativa, essa observação pode ser confirmado mediante valor calculado correlação= -0.98, o que sigifica que os dados são fortemente correlacionados, dessa forma a regressão linear foi uma estratégia adequada.

## Resposta 4a
Neste caso é necessário definir um diagrama de dispersão considerando o início em (0,0), usaremos uma alternativa para realizar o gráfico marcando as linhas tracejadas (Obs. No programa MATLAB esta ferramenta se conhece como grid)

```python
# definir dados
x = [59, 63, 68, 72, 74, 78, 83] # Temperatura [°C]
y = [118, 182, 247, 208, 197, 135, 132] #  Aumento de % ao derreter o queijo
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Delimitar o gráfico iniciando em (0,0)
plt.xlim(0,105)   #axis x 
plt.ylim(0,260)   #axis y
# Legendas dos eixos
plt.xlabel('Temperatura [°C] ')
plt.ylabel('Aumento de % ao derreter o queijo ')  
# Marcar os pontos indicados na questão 4a 
x1= [0, 20, 40, 60, 80, 100]
y1= [0, 50, 100, 150, 200, 250]
# Plotar o conjunto de dados obtidos no experimento
plt.scatter(x,y, c='green', label='Dados experimentais') 
# Plotar os dados indicados na questão
plt.scatter(x1,y1, c='coral', label='Dados solicitados')
plt.legend()
plt.grid(True)
plt.show()
```

No resultado do gráfico pode-se observar que os dados em cor coral (solicitados no item 4a) apresentam o comportamento de uma reta com inclinação de 2,5 (ou 5/2) e intercepto de 0. Já os dados do experimento de temperatura de derretimento do queijo (em verde) se afastam deste tipo de comportamento.

## Resposta 4b
Agora o diagrama de dispersão pode ser restringido ao intervalo onde os dados aparecem, overse que o menor valor de x é 59 e para y é 118, e os maiores valores de x e y são 83 e 247 respectivamente. Portanto os pontos que delimitam o novo gráfico de dispersão devem ser (59,118) e (83,247). para fins de observação podemos delimitar o gráfico com os seguintes pontos (55, 100), como solicitado neste item, e (90, 250)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Delimitar o gráfico com os pontos (55, 100) e (90, 250)
plt.xlim(55,90)   #axis x 
plt.ylim(100,250)   #axis y
# Legendas dos eixos
plt.xlabel('Temperatura [°C] ')
plt.ylabel('Aumento de % ao derreter o queijo ')  
# Plotar o conjunto de dados obtidos no experimento calculando o coeficiente de correlação
plt.scatter(x, y, c='green', label=f'Correlação = {np.round(np.corrcoef(x,y)[0,1], 2)}')
plt.grid(True)
plt.legend()
plt.show()
```

Observando este gráfico nada muda quando comparado com o item anterior, pois apesar de ter reduzido os eixos ele continua apresentando um comportametno afastado da linha reta. Essa afirmação pode ser comprovada observando o valor do coeficiente de correlação próximo de zero.
