```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
from statsmodels.graphics.gofplots import ProbPlot
plt.style.use('seaborn')
plt.rc('axes', titlesize=10)
```

```python
x = [1326, 1593, 4414, 5673, 29516, 26, 843, 1016, 3410, 7101, 7356, 7904, 79, 4175, 34676, 114789, 2672, 7532, 30220]
y = [0.01495, 0.01470, 0.01100, 0.01190, 0.00873, 0.01819, 0.00810, 0.0801, 0.00600, 0.00575, 0.00576, 0.00580, 0.01212, 0.00782, 0.00596, 0.00600, 0.00880, 0.00883, 0.00676]
```

```python
plt.scatter(x,y)
```

```python
vtx=np.log(x)
plt.scatter(vtx,y)
# amplitude do esforço dos ciclos em motores de avião até uma falha ocorrer
plt.xlabel('Ciclos até a falha')
plt.ylabel('Amplitude esforço')
plt.title('Gráfico com variáveis x e y transformadas')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame(list(zip(x, y, vtx)),
               columns =["x","y","ln_x"])
df
```

```python
vtx = sm.add_constant(vtx)
#fit linear regression model
modelo = sm.OLS(y, vtx).fit() #atenção esta variável será usada no IC
print(modelo.summary())
```

```python
# parâmetros da reta
b=modelo.params
b0=np.round(b[0], 4) #intercepto
b1=np.round(b[1], 4) #inclinação
print('y= {0}{1} ln_x'.format(b0,b1))
```

```python
x_def = 2710
y_esp = b0 + b1*np.log(x_def)
y_esp
```
